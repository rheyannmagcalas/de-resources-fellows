```python
import pandas as pd

from pandasql import sqldf, load_meat, load_births
pysqldf = lambda q: sqldf(q, globals())
```

```python
meat = load_meat()
meat
```

```python
births = load_births()
births
```

```python
pysqldf("SELECT * FROM meat LIMIT 10;")
```

```python
q = """SELECT
        m.date, m.beef, b.births
     FROM
        meat m
     INNER JOIN
        births b
           ON m.date = b.date;"""
joined = pysqldf(q)
joined.head()
```

```python
q = "SELECT strftime('%Y', date) AS year, SUM(beef) AS beef_total FROM meat GROUP BY year;"
pysqldf(q)
```

```python
titanic_data = pd.read_csv('data/titanic.csv')
titanic_data
```

```python
q = "SELECT * from titanic_data where Age > 50 "
pysqldf(q)
```

```python
q = "SELECT * from titanic_data where Age > 50 group by Sex"
pysqldf(q)
```
